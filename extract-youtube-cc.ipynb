```python
from youtube_transcript_api import YouTubeTranscriptApi

#https://www.youtube.com/watch?v=knW7-x7Y7RE

video_id = "knW7-x7Y7RE"
# video_id = "Rf09BOJgObE"

try:
    # Retrieve the transcript (v1.0+ API)
    ytt_api = YouTubeTranscriptApi()
    transcript = ytt_api.fetch(video_id, languages=['vi', 'en'])

    # Print the first 3 lines
    for entry in list(transcript):
        print(f"{entry.start}: {entry.text}")

except Exception as e:
    print(f"Error: {e}")
```

```python
try:
    ytt = YouTubeTranscriptApi()
    transcript_list = ytt.list(video_id)

    print("Co transcript.")
    print("Danh sach ngon ngu:")
    for t in transcript_list:
        # t.language_code: ma ngon ngu, t.language: ten ngon ngu
        print(f"- {t.language_code}: {t.language} | generated={t.is_generated}")

except Exception as e:
    print(f"Khong lay duoc transcript: {e}")
```
