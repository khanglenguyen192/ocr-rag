```python
# Cell 1: Cài thư viện (chạy một lần)
# !pip install pymupdf4llm

# Cell 2: Import & cấu hình
import pymupdf4llm
from pathlib import Path

INPUT_DIR  = Path("input")
OUTPUT_DIR = Path("output")
OUTPUT_DIR.mkdir(exist_ok=True)

# Cell 3: Convert một file PDF cụ thể
TARGET_FILE = INPUT_DIR / "chess-rule.pdf"   # Thay đổi tên file tuỳ ý

# Tạo thư mục chứa ảnh riêng cho từng file
img_dir = OUTPUT_DIR / TARGET_FILE.stem
img_dir.mkdir(exist_ok=True)

md_text = pymupdf4llm.to_markdown(
    str(TARGET_FILE),
    write_images=True,          # Giữ lại ảnh
    image_path=str(img_dir),    # Thư mục lưu ảnh
    image_format="png",         # Định dạng ảnh (png / jpeg)
)

out_path = OUTPUT_DIR / (TARGET_FILE.stem + ".md")
out_path.write_text(md_text, encoding="utf-8")

print(f"Đã lưu: {out_path}")
print(f"Ảnh lưu tại: {img_dir}")
print(f"Số ký tự: {len(md_text)}")
```

```python
# Cell 5: Convert toàn bộ PDF trong thư mục input/
results = {}
for pdf in sorted(INPUT_DIR.glob("*.pdf")):
    try:
        img_dir = OUTPUT_DIR / pdf.stem
        img_dir.mkdir(exist_ok=True)
        text = pymupdf4llm.to_markdown(
            str(pdf),
            write_images=True,
            image_path=str(img_dir),
            image_format="png",
        )
        out  = OUTPUT_DIR / (pdf.stem + ".md")
        out.write_text(text, encoding="utf-8")
        results[pdf.name] = str(out)
    except Exception as e:
        results[pdf.name] = f"ERROR: {e}"

print("── Kết quả ──")
for src, dst in results.items():
    print(f"  {src}  →  {dst}")
```
